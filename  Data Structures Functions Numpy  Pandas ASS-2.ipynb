### 7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
num = int(input("enter the number"))

if num > 0:
    print("The number is positive.")
    if num % 2 == 0:
        print("It is also even.")
    else:
        print("It is odd.")
        
elif num == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```

### 8.	Create:

### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:

```python
import pandas as pd
import numpy as np
```

```python
data = {
        'Name'   : ['Snehdeep','Priyanshi','Aashu','Senhdeep','Senhyansh','Snehyanshii','Manas','Jay','Aakash','Rohit'],
        'Age'    : [23,22,24,23,23,28,25,27,28,29],
        'City'   : ['Bengaluru', 'Gwalior', 'Mumbai','Bengaluru', 'Chennai','Bhopal', 'Kolkata', 'Jaipur', 'Delhi', 'Pune'],
        'Salary' : [500000, 60000, np.nan, 500000, 70000, 40000, 65000, np.nan, 72000, 58000]
}
df = pd.DataFrame(data)
df
```

```python
df.loc[:,['Name','City']]
```

```python
mean_salary = df['Salary'].mean()
mean_salary
```
