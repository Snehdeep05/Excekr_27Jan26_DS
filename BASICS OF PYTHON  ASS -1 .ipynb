## Exercise 1: Prime Numbers


#### Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
# user input
num = int(input("Enter a number: "))

#conditional statment
if num <= 1:
    print("Not a prime number")
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

## Exercise 2: Product of Random Numbers

#### Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

```python
# generate two random numbers
import random
a = random.randint(1, 8)
b = random.randint(1, 8)

print("Number 1:", a)
print("Number 2:", b)

# asking from the user answer
user_answer = int(input("Enter the product: "))

# products of two random numbers 
correct_answer = a * b

# check
if user_answer == correct_answer:
    print("Correct! Well done.")
else:
    print("Wrong answer.")
    print("Correct product is:", correct_answer)
```

## Exercise 4: Problem Statement (Question): Hospital Billing System

### A hospital charges its patients based on the following:
#####	Room charges per day: ₹2,000
#####	Doctor consultation fee (flat): ₹1,500
#####	Lab test charges: Based on the number of tests taken. Each test costs ₹300
#####	Medicine charges: Total cost of medicines provided
#####	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
###    Write a Python program to calculate the final bill for a patient who:
#####	Stayed for 4 days
#####	Had 3 lab tests
#####	Had medicine charges of ₹2,400



```python
# Fixed charges
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_cost = 300

# Given
days = 5

# User inputs
tests = int(input("Enter number of lab tests: "))
medicine_cost = float(input("Enter medicine charges: ₹"))

# Calculations
room_total = days * room_charge_per_day
lab_total = tests * lab_test_cost

total_bill = room_total + doctor_fee + lab_total + medicine_cost

# Discount
if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount

# Output
print("\n--- Hospital Bill ---")
print("Room charges:", room_total)
print("Doctor fee:", doctor_fee)
print("Lab charges:", lab_total)
print("Medicine charges:", medicine_cost)
print("Total before discount:", total_bill)
print("Discount:", discount)
print("Final bill:", final_bill)
```
